```python
def surface(S):
    I=S.first_fundamental_form_coefficients()
    matrixI=matrix(2,[I[(1,1)].simplify_full(),I[(1,2)].simplify_full(),I[(2,1)].simplify_full(),I[(2,2)].simplify_full()])
    II=S.second_fundamental_form_coefficients()
    matrixII=matrix(2,[II[(1,1)].simplify_full(),II[(1,2)].simplify_full(),II[(2,1)].simplify_full(),II[(2,2)].simplify_full()])
    
    show(matrixI)
    show(matrixII)
    show(S.gauss_curvature().factor())
```

If $c = 1$, we have the same surface, so the isometry is clear.

If $c > 1$, we have an isometry from $U:=\varphi((-\arcsin(1/c),\arcsin(1/c))\times(0,2\pi))$ (on the surface S3) to $V:=\varphi((-\arcsin(1/c),\arcsin(1/c))\times (0,2\pi))$ (on the sphere) because note that the two matrices coincide everywhere they are defined.

```python
# For c = 2 we have:
p1=S41.plot((T2[0],T2[1]),(0,2*pi),color='red',aspect_ratio=1)
p2=S42.plot((T2[0],T2[1]),(0,2*pi),color='green',aspect_ratio=1)
p1+p2
```

If $c < 1$, we have an isometry from $U:=\varphi((-\pi/2,\pi/2)\times(0,2\pi))$ (on the surface S3) to $V:=\varphi((-\pi/2,\pi/2)\times (0,2\pi))$ (on the sphere) because note that the two matrices coincide everywhere on the whole interval where they are defined.

```python
# For c = 1 / 2 we have:
p3=S41.plot((T3[0],T3[1]),(0,2*pi),color='red',aspect_ratio=1,opacity=0.7)
p4=S43.plot((T3[0],T3[1]),(0,2*pi),color='blue',aspect_ratio=1)
p3+p4
```
